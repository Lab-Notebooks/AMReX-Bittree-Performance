```python
import matplotlib.pyplot as pyplot
```

```python
# Plot figure
pyplot.rc("font", family="serif", size=14, weight="bold")
pyplot.rc("axes", labelweight="bold", titleweight="bold")
pyplot.rc("text", usetex=True)
figure = pyplot.figure(figsize=(5, 5), dpi=80)

procList = [168, 2688, 21504]

# Create subplots
ax1 = figure.subplots(1,1)

ax1.plot(procList, [35.109, 49.072, 85.057], marker="o", markersize=12)
ax1.plot(procList, [33.686, 50.504, 189.707], marker="s", markersize=12)
ax1.plot(procList, [104.692, 104.593, 107.639], marker="d", markersize=12)

ax1.set_title("Regridding Time in Flash-X")
ax1.legend(["AMReX Bittree", "AMReX Native", "Paramesh Bittree"])
ax1.set_xlabel(r"Ranks")
ax1.set_ylabel(r"Time (s)")
ax1.set_yscale("log")
ax1.set_yticks([10, 100, 1000])
pyplot.tight_layout()
pyplot.savefig("/Users/Akash/Desktop/bittree.png")
```
